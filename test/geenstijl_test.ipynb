```python
import pandas as pd
import re 

df = pd.read_json('../output/news.json')
df.head()
```

```python
# apply date_func and time_func on "footer" column to seprate article time and date and store them in seprate columns 
def date_func(text):
    date = re.findall(r'[0-9]{2}[-|\/]{1}[0-9]{2}[-|\/]{1}[0-9]{2}', text)
    return date
def time_func(text):
        time = re.findall(r'[0-9]{2}[:|\/][0-9]{2}', text)
        return time
    
df = df.replace('\n','', regex=True)
df['publication_date']= df['footer'].apply(lambda x : date_func(x))
df['publication_time']= df['footer'].apply(lambda x : time_func(x))
df.head(2)
```
